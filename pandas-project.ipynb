```python
import pandas as pd
import os

filename = "adult_data_cleaned.csv"

# Check if the file already exists on your computer
if os.path.exists(filename):
    df = pd.read_csv(filename)
    print("✅ Loaded from local storage (Fast!)")
else:
    # If not, download it from UCI
    url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
    columns = ["age", "workclass", "fnlwgt", "education", "education_num",
               "marital_status", "occupation", "relationship", "race", "gender",
               "capital_gain", "capital_loss", "hours_per_week", "native_country", "income"]
    df = pd.read_csv(url, names=columns, skipinitialspace=True)
    # Save it for next time
    df.to_csv(filename, index=False)
    print("🌐 Downloaded from UCI and saved locally.")
```

```python
import numpy as np
#replace '?' with standar null value(NAN)
df.replace('?',np.nan,inplace=True)
#drop rows that having missing values
df.dropna(inplace=True)
print(f"✅ Data cleaned. Remaining rows: {len(df)}")
```

```python
# 1. Import the tool from the Scikit-Learn library
from sklearn.preprocessing import LabelEncoder

# 2. Create an instance of the encoder
le = LabelEncoder()

# 3. Define which columns are "Categorical" (Text-based)
categorical_cols = ["workclass", "education", "marital_status", "occupation", 
                    "relationship", "race", "gender", "native_country", "income"]

# 4. Use a loop to transform each column
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])

# 5. Check the result
print("✅ Encoding complete!")
df.head()
```
